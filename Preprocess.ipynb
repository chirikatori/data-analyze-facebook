## Data preprocessing

```python
data = raw_df.dropna(axis="columns", thresh=100)
data = raw_df.dropna(subset="w3_fb_url")
```

```python
data.info()
```

### Drop columns that is not necessory

```python
data = data.drop_duplicates(subset="post_id")
data = data.drop(columns=["was_live", "fetched_time", 
                           "timestamp", "image", "images", 
                           "timestamp", "image", "image_lowquality",
                          "images","images_description", 
                          "images_lowquality_description",
                          "is_live", "w3_fb_url", "user_id", "username", "user_url",
                          "available", "image_id", "image_ids", "post_url"
                          ])


data.head()

```

### Create column "has_video" and drop all video columns

```python
has_vid = []
for vid in data["video"]:
    if vid is not np.nan:
        has_vid.append(True)
    else:
        has_vid.append(False)
data["has_vid"] = pd.DataFrame(has_vid)
data.head()
```

```python
data = data.drop(columns=["video", "video_id", "video_thumbnail"])

data.head()
```

### Create column "has_image"

```python
has_imp = []
for image in data["images_lowquality"]:
    if image is not np.nan:
        has_imp.append(True)
    else:
        has_imp.append(False)
data["has_image"] = pd.DataFrame(has_imp)
data.head()
```

```python
data = data.drop(columns="images_lowquality")
data.reset_index()
```

```python
data.head()
```

### Make a columns name "post_time" to show the frequency of time page post a post

```python
convert_to_time = []
for time in data["time"]:
    try:
        convert_to_time.append(datetime.strptime(time, '%Y-%m-%d %H:%M:%S'))
    except:
        convert_to_time.append(np.nan)
len(convert_to_time)

data["post_time"] = pd.DataFrame(convert_to_time)
data["post_time"] = pd.to_datetime(data["post_time"]).dt.time
```

```python
data = data.dropna(axis=0, subset=["post_time"])
```

## Make type of reaction columns

```python
reaction = data["reactions"].apply(ast.literal_eval).apply(pd.Series)
reaction
```

### Convert NaN value into 0 and float into integer

```python
reaction = reaction.fillna(0).astype(int)
reaction
```

### Add reaction Dataframe into data

```python
data = pd.concat([data, reaction], axis=1)
data
```

## Save clean file

```python
data.to_csv(f"{os.getenv('FOLDER_PATH_CLEAN')}{os.getenv('FACEBOOK_FANPAGE_ID')}.csv", mode="w", index=False)
```
